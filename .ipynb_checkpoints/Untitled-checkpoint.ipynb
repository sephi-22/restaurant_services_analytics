# Cleaning Start

Let's go from the smallest Dataset to the Biggest. The Scraped information CSV is going to be my starting point. I will transform other datasets to fit in line with this dataset

```python
ahistoric_df.head()
```

```python
def decode_html_entities(column):
    return column.apply(lambda x: html.unescape(str(x)))
```

```python
def capitalize_column(dataframe, column):
    dataframe[column]=dataframe[column].str.title()
    return dataframe
```

```python
ahistoric_df['Restaurant Name'] = decode_html_entities(ahistoric_df['Restaurant Name'])
ahistoric_df['Doing Business As (DBA)'] = decode_html_entities(ahistoric_df['Doing Business As (DBA)'])
```

```python
ahistoric_df.head()
```

```python
ahistoric_df['Borough'].value_counts()
```

```python
ahistoric_df['Postcode'].value_counts()
```

```python
ahistoric_df['Building Number'].value_counts()
```

```python
ahistoric_df = capitalize_column(ahistoric_df,'Restaurant Name')
ahistoric_df = capitalize_column(ahistoric_df,'Doing Business As (DBA)')
ahistoric_df = capitalize_column(ahistoric_df,'Street')
ahistoric_df = capitalize_column(ahistoric_df,'Legal Business Name')


ahistoric_df.head()
```

```python
ahistoric_df
```

```python
replacement_dict = {'Llc': 'LLC', 'Inc': 'Inc.', 'Corp':'Corp.'}
```

```python
for term, replacement in replacement_dict.items():
    ahistoric_df['Legal Business Name'] = ahistoric_df['Legal Business Name'].str.replace(term, replacement)
    ahistoric_df['Doing Business As (DBA)'] = ahistoric_df['Doing Business As (DBA)'].str.replace(term, replacement)
```
