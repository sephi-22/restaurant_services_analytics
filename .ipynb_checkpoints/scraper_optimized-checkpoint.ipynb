```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
from bs4 import BeautifulSoup
import time
import pandas as pd


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

main_list = []
base_url = "http://www.yellowpages.com"
number_of_pages_to_scrape = 100
num_iterations = 30

for page_number in range(1, number_of_pages_to_scrape + 1):
    url = f'https://www.yellowpages.com/search?geo_location_terms=New%20York%20City%2C%20NY&search_terms=restaurants&page={page_number}'
    source = requests.get(url, headers=headers,timeout =120).text
    soup = BeautifulSoup(source, 'html.parser')

    all_info_tags = soup.find_all('div', class_='info')
    
    print()
    print(page_number)
    print()

    for value in all_info_tags:
# for index, value in enumerate(all_info_tags[:num_iterations], start=1)
        # restaurent name
        restaurant_name = value.find('a', class_='business-name').text  # getting the restaurant name from the info_div

        # categories
        category_div =[category for category in value.find_all('div',class_='categories')] # finding the category class from the info_div
        category_a_tags = [item for sublist in category_div for item in sublist]
        categories = [item.text.strip() for item in category_a_tags]

        # trip adviser ratings
        try:
            ta_ratings = value.find('div',class_='ratings')['data-tripadvisor']  
        except Exception as e:
            ta_ratings = None 

        # yellow pages ratings and count
        try:
            y_rating = value.find('div',class_='ratings').a.div['class'][1:]
            if len(y_rating) != 1:
                yellow_ratings = " and ".join(y_rating)
            else:
                yellow_ratings = " ".join(y_rating)
        except:
            y_rating = None
            yellow_ratings = None

        try:
            y_rating_count = value.find('div',class_='ratings').a.span.text.strip('()')
        except:
            y_rating_count = None

        yellow_ratings_and_count = {'rating':yellow_ratings,'count':y_rating_count}

        # links
        base_url = "http://www.yellowpages.com"
        website_link = value.find('div',class_='links')
        try:
            website = website_link.find('a',class_='track-visit-website')['href']
        except:
            website = None

        try:
            directions = website_link.find('a',class_='track-map-it directions')['href']
            direction_website = base_url + directions
        except:
            directions = None
            direction_website = None

        try:
            menu = website_link.find('a',class_='menu')['href']
            menu_website = base_url+menu
        except:
            menu = None
            menu_website = None

        # Total business years
        try:
            total_business_years = value.find('div',class_='count').text 
        except:
            total_business_years = None

        # Amenities
        try:
            amenities = value.find('div',class_='amenities-info')
            amenities_available = [i.text for i in amenities.find_all('span')]
        except:
            amenities = None
            amenities_available = None


        # description    
        try:
            description = value.find('div',class_='snippet').p.text
        except:
            description = None

        # Order online status
        online = value.find('div',class_='listing-ctas').find('a',class_='action order-online')
        status = True if online else False

        # phone number
        try:
            phone_number = value.find('div',class_='phones phone primary').text
        except:
            phone_number = None

        # street address
        try:
            street = value.find('div',class_='street-address').text
        except:
            street = None

        # loaclity
        try:
            locality = value.find('div',class_='locality').text
        except:
            locality = None

        # price range
        try:
            price_range = value.find('div',class_='price-range').text
        except:
            price_range = None

        data = {'Name':restaurant_name,'Type':categories,'Trip_advisor_ratings':ta_ratings,'Yellow_page_ratings':yellow_ratings_and_count,'Website_link':website,'Directions_link':direction_website,'Menu_link':menu_website,'Total_business_years':total_business_years,'Amenities_available':amenities_available,'Description':description,'Order_online':status,'Phone_Number':phone_number,'Street_address':street,"locality":locality,"Price_range":price_range}
        print(data)
        main_list.append(data)
        print()


```
