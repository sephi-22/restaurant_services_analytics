```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import time
```

```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                    'Chrome/101.0.4951.54 Safari/537.36',
        'Connection': 'keep-alive',
        'Referer': 'https://google.com',
        'DNT': '1',
        'Accept-Language': 'en-GB,en;q=0.5'
    }
```

```python
url = "https://www.yellowpages.com/search?search_terms=restaurants&geo_location_terms=New+York+City%2C+NY"
```

```python
field_names = ['Name','Type of Restaurant','Trip Advisor Rating','TA Review Count','Yellow Pages Rating','YP Review Count','Phone Number','Address','Dollar Costs','Years in Business','Website','Menu URL','Secondary Info','Order Online Status']
```

```python
current_page = url
while current_page:
    response = requests.get(current_page, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text,'html.parser')
        for result in soup.find_all('div', class_='info'):
            # Extract name
            name_text = result.find('a', class_='business-name').find('span').text.strip()
            print(name_text)
        
            # Extract categories
            category_div = result.find('div',class_='categories')
            if category_div:
                category_elements = category_div.find_all('a')
                categories = [category.text.strip() for category in category_elements]
            else:
                categories = None
            print("Categories:", categories)
        
            # Extract Tripadvisor Ratings
            ta_rating_data = result.find('div', class_='ratings')['data-tripadvisor']
            if ta_rating_data:
                ta_rating = float(ta_rating_data.split('"rating":"')[1].split('"')[0])
                ta_rating_count = int(ta_rating_data.split('"count":"')[1].split('"')[0])
            else:
                ta_rating = None
                ta_rating_count = None
            print("Trip Advisor Rating:", ta_rating)
            print("Trip Advisor Count:", ta_rating_count)
            
             # Extract Yellowpages Ratings
            has_extra_rating = result.find('a', class_='hasExtraRating')
            if has_extra_rating:
                yp_rating_class = has_extra_rating.find('div', class_='result-rating')['class'][1:]
                yp_rating_count = int(has_extra_rating.find('span', class_='count').text.strip('()'))
            else:
                yp_rating_class = None
                yp_rating_count=None
            print("Yellow Pages Rating:", yp_rating_class)
            print("Yellow Pages Count:", yp_rating_count)
            


        # Extract phone
            phone = result.find('div', class_='phones')
            if phone:
                phone_text = phone.text.strip()
                print("Phone:",phone_text)
            else:
                print("Phone not found")
            

        # Extract address
            address = result.find('div', class_='adr')
            if address:
                address_text = address.text.strip()
                print("Address:",address_text)
            else:
                print("Address not found")

        # Extract open status
            open_status = result.find('div', class_='open-status')
            if open_status:
                open_status_text = open_status.text.strip()
                print("Open Status:",open_status_text)
            else:
                print("Open Status not found")


        # Print or store the extracted information
      
```
